```python
import nltk
from nltk import *
sent = "India is a republic nation. We are the proud nation"
print(len(sent))
print(sent[0:5])
print(sent[11:19])
nltk.download('punkt_tab')
print(nltk.word_tokenize(sent))

tokens = nltk.word_tokenize(sent)
vocab = sorted(set(tokens))

from string import punctuation
vocab_wo_punct = []
for i in vocab:
    if i not in punctuation:
        vocab_wo_punct.append(i)
print(vocab_wo_punct)

from nltk import pos_tag
nltk.download('averaged_perceptron_tagger_eng')
pos_list = pos_tag(vocab_wo_punct)
print(pos_list)


from nltk.corpus import stopwords
nltk.download('stopwords')
wo_stop_words = []
stopwords_set = set(stopwords.words("english"))
for i in vocab_wo_punct:
  if i not in stopwords_set:
    wo_stop_words.append(i)
print(wo_stop_words)



from nltk.stem.snowball import SnowballStemmer
stemObj = SnowballStemmer('english')
stemObj.stem('studying')

stemed_vocab =[]
stemObj = SnowballStemmer('english')
for i in vocab_wo_punct:
    stemed_vocab.append(stemObj.stem(i))
print(stemed_vocab)

from nltk.stem.wordnet import WordNetLemmatizer
nltk.download('wordnet')
lemmaObj = WordNetLemmatizer()
lemmaObj.lemmatize("went",pos='v')

```

```python
# Frequency distribution
text ="I saw John coming. He was with Mary. I talked to John and Mary. John said he met Mary on the way. John and Mary were going to school "
print(nltk.FreqDist(nltk.word_tokenize(text)))
nltk.FreqDist(text.split()).plot()


# Conditional Frequency Distribution
from nltk.corpus import brown
import nltk
nltk.download('brown')
nltk.download('punkt')

cfd = nltk.ConditionalFreqDist(
          (genre, word)
          for genre in brown.categories()
          for word in brown.words(categories=genre))
genres_list = ['romance','news','science_fiction', 'humor','religion','hobbies']
modals_list = ['may','could', 'can', 'must', 'will', 'might']
cfd.tabulate(conditions=genres_list, samples=modals_list)



# N-gram
from nltk import ngrams
from nltk.util import bigrams, trigrams
#use 2 for bigrams
vocab_no_punct = "India Indians We a are country is proud republic".split()
bigrams = ngrams(vocab_no_punct,2)
print(list(bigrams))
#Prints [('India', 'Indians'), ('Indians', 'We'), ('We', 'a'), ('a', 'are'), ('are', 'country'),
        #  ('country', 'is'), ('is', 'proud'), ('proud', 'republic')]
#use 3 for trigrams
trigrams = ngrams(vocab_no_punct,3)
print(list(trigrams))
# [('India', 'Indians', 'We'), ('Indians', 'We', 'a'), ('We', 'a', 'are'), ('a', 'are', 'country'),
# ('are', 'country', 'is'), ('country', 'is', 'proud'), ('is', 'proud', 'republic')]
```
